# The percentage of NTB customers who onboarded via Alat, USSD, or AFB?

# The average age of NTB customers based on their date of birth.

```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime

# Define the date format (adjust this format based on your data)
date_format = "%Y-%m-%d"

# Convert the 'Date_of_Birth' column to datetime with the specified format
demographic_data_clean['Date_of_Birth'] = pd.to_datetime(demographic_data_clean['Date_of_Birth'], format=date_format, errors='coerce')

# Calculate the age of customers based on their date of birth
current_year = datetime.datetime.now().year
demographic_data_clean['Age'] = current_year - demographic_data_clean['Date_of_Birth'].dt.year

# Calculate the average age
average_age = demographic_data_clean['Age'].mean()

# Set the figure size
plt.figure(figsize=(8, 6))

# Create a histogram to visualize the distribution of ages
plt.hist(demographic_data_clean['Age'], bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.title(f'Distribution of Customer Ages\n(Average Age: {average_age:.2f} years)')

# Show the plot
plt.tight_layout()
plt.show()
```

# Account Activity and Balances:- What is the average account balance for different customer segments (e.g., gender, age, location)?

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Calculate the age of customers based on their date of birth
current_year = pd.Timestamp.now().year
demographic_data_clean['Date_of_Birth'] = pd.to_datetime(demographic_data_clean['Date_of_Birth'], errors='coerce')
demographic_data_clean['Age'] = current_year - demographic_data_clean['Date_of_Birth'].dt.year

# Group the data by different customer segments and calculate the average account balance
average_balance_by_gender = demographic_data_clean.groupby('Gender')['Account_Balance'].mean()
average_balance_by_age_group = demographic_data_clean.groupby(pd.cut(demographic_data_clean['Age'], bins=[0, 20, 30, 40, 50, 60, float('inf')]))['Account_Balance'].mean()
average_balance_by_location = demographic_data_clean.groupby('Location')['Account_Balance'].mean()

# Create subplots for each segment
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Plot average account balance by gender
axes[0].bar(average_balance_by_gender.index, average_balance_by_gender.values, color='skyblue')
axes[0].set_ylabel('Average Account Balance')
axes[0].set_title('Average Account Balance by Gender')

# Plot average account balance by age group
average_balance_by_age_group.plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_ylabel('Average Account Balance')
axes[1].set_title('Average Account Balance by Age Group')
axes[1].set_xlabel('Age Group')

# Plot average account balance by location
average_balance_by_location.plot(kind='bar', ax=axes[2], color='lightcoral')
axes[2].set_ylabel('Average Account Balance')
axes[2].set_title('Average Account Balance by Location')
axes[2].set_xlabel('Location')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the sample size
sample_size = 500

# Randomly sample from both demographic_data_clean and transactional_data_clean
demographic_sample = demographic_data_clean.sample(sample_size, random_state=42)
transactional_sample = transactional_data_clean.sample(sample_size, random_state=42)

# Convert 'Transaction_Date' column to datetime in transactional_sample
transactional_sample['Transaction_Date'] = pd.to_datetime(transactional_sample['Transaction_Date'])

# Merge the two samples based on 'Account_Number'
merged_sample = demographic_sample.merge(transactional_sample[['Account_Number', 'Transaction_Date']], on='Account_Number', how='left')

# Identify NTB customers who haven't made any transactions
merged_sample['Active'] = merged_sample['Transaction_Date'].isnull()

# Group the data by date and count the number of NTB customers who became active/transacting on each date
conversion_data = merged_sample[merged_sample['Active'] == False].groupby('Transaction_Date').size().reset_index(name='Count')

# Create a line plot to visualize the conversion trend over time
plt.figure(figsize=(12, 6))
plt.plot(conversion_data['Transaction_Date'], conversion_data['Count'], marker='o', linestyle='-', color='blue')
plt.xlabel('Transaction Date')
plt.ylabel('Number of NTB Customers Becoming Active')
plt.title('Conversion of NTB Customers into Active Users Over Time')
plt.grid(True)
plt.show()
```
